Tôi muốn quay lại một chút và chỉ ra rằng bạn có thể đã thấy, trong phân phối dữ liệu này về các lớp khác nhau của tập dữ liệu đã chọn cuối cùng, bạn có thể thấy rằng "Happy" có rất nhiều mẫu trong khi "Disgust" chỉ có 6 mẫu. Điều này đặt ra một vấn đề trong việc chia tập huấn luyện và tập kiểm tra. Nếu bạn nghĩ về việc chia tập dữ liệu huấn luyện và kiểm tra theo tỷ lệ thông thường, ví dụ như 70-30%, tức là 70% dữ liệu dùng cho huấn luyện và 30% dùng cho kiểm tra, thì có thể xảy ra trường hợp một số lớp như "Disgust" sẽ không xuất hiện trong cả hai tập dữ liệu. Điều này có thể dẫn đến việc mô hình không thể học cách dự đoán lớp "Disgust" một cách hiệu quả, và điều này sẽ gây khó khăn trong quá trình huấn luyện.

Một cách giải quyết là chúng ta có thể chia các lớp dựa trên tập huấn luyện và tập kiểm tra một cách cân bằng. Điều này được gọi là phương pháp chia tập huấn luyện và kiểm tra theo lớp (stratified approach).

Trong nhiệm vụ này, chúng ta sẽ sử dụng hàm train_test_split từ thư viện Scikit-Learn để thực hiện chia tập huấn luyện và tập kiểm tra theo cách cân bằng. Chúng ta sẽ chia dựa trên chỉ số và nhãn, với tỷ lệ tập kiểm tra là 15% và 85% dành cho tập huấn luyện. Sau đó, chúng ta sẽ tạo một cột mới trong bộ dữ liệu để xác định loại dữ liệu (huấn luyện hoặc kiểm tra) mà mỗi mẫu thuộc về.

## Task 4: Loading Tokenizer and Encoding our Data

```python
from transformers import BertTokenizer
from torch.utils.data import TensorDataset
```

```python
tokenizer = BertTokenizer.from_pretrained(
    'bert-base-uncased',
    do_lower_case = True
)
```

* `add_special_tokens`: True, để thêm các mã thông báo đặc biệt (như mã thông báo [CLS] và [SEP]) vào các chuỗi văn bản.

* `return_attention_mask`: True, để trả về ma trận attention mask, mà là một ma trận có cùng kích thước với ma trận input_ids, nhưng có giá trị 0 tại vị trí các mã thông báo pad và 1 tại các vị trí có dữ liệu thực sự.

* `pad_to_max_length`: True, để đảm bảo tất cả các chuỗi đều có cùng độ dài bằng cách thêm mã thông báo pad.

* `max_length`: 256, là độ dài tối đa của các chuỗi sau khi mã hóa. Các chuỗi dài hơn sẽ bị cắt bớt, còn ngắn hơn sẽ được điền vào bằng pad tokens.

* `return_tensors`: "pt", để trả về các tensors PyTorch.

```python
# Định nghĩa tham số chung cho mã hóa
encode_params = {
    "add_special_tokens": True,
    "return_attention_mask": True,
    "padding": "max_length",
    "truncation": True,
    "max_length": 256,
    "return_tensors": "pt"
}

# Mã hóa dữ liệu huấn luyện
encoded_data_train = tokenizer.batch_encode_plus(
    df[df.data_type == 'train'].text.values,
    **encode_params
)
input_ids_train = encoded_data_train['input_ids']
attention_masks_train = encoded_data_train['attention_mask']
labels_train = torch.tensor(df[df.data_type == 'train'].label.values)

# Mã hóa dữ liệu thử nghiệm
encoded_data_val = tokenizer.batch_encode_plus(
    df[df.data_type == 'val'].text.values,
    **encode_params
)
input_ids_val = encoded_data_val['input_ids']
attention_masks_val = encoded_data_val['attention_mask']
labels_val = torch.tensor(df[df.data_type == 'val'].label.values)
```

```python
dataset_train = TensorDataset(input_ids_train,
                              attention_masks_train, labels_train)
dataset_val = TensorDataset(input_ids_val,
                            attention_masks_val, labels_val)
```

## Task 5: Setting up BERT Pretrained Model

```python
from transformers import BertForSequenceClassification
```

* `BertForSequenceClassification`: Đây là lớp mô hình của BERT được tạo ra để thực hiện bài toán phân loại chuỗi văn bản.

* `.from_pretrained(...)`: Phương thức tĩnh của lớp BertForSequenceClassification để tạo mô hình từ các trọng số của mô hình đã được huấn luyện trước.

* `'bert-base-uncased'`: Đây là tên của mô hình BERT cơ sở (base model) chưa được chuyển thành chữ hoa, có nghĩa là các từ đã được viết thường hết. Mô hình này đã được huấn luyện sẵn trên dữ liệu lớn và có sẵn trong thư viện Transformers.

* `num_labels = len(label_dict)`: Đây là tham số num_labels để xác định số lượng nhãn (classes) mà mô hình sẽ dự đoán. Giá trị của num_labels là độ dài của từ điển nhãn (label_dict), tức là số lớp bạn đang dự đoán.

* `output_attention = False`: Tham số này xác định liệu bạn có muốn lấy ma trận attention (ma trận chú ý) của mô hình ra hay không. Trong trường hợp này, bạn đã đặt False, có nghĩa là không cần lấy ra ma trận attention.

* `output_hidden_states = False`: Tham số này xác định liệu bạn có muốn lấy ra các trạng thái ẩn (hidden states) của mô hình hay không. Ở đây, bạn cũng đã đặt False, có nghĩa là không cần lấy ra các trạng thái ẩn.

```python
model = BertForSequenceClassification.from_pretrained(
    'bert-base-uncased',
    num_labels = len(label_dict),
    output_attentions = False,
    output_hidden_states = False
)
```

## Task 7: Setting Up Optimizer and Scheduler

## Task 8: Defining our Performance Metrics

Accuracy metric approach originally used in accuracy function in [this tutorial](https://mccormickml.com/2019/07/22/BERT-fine-tuning/#41-bertforsequenceclassification).

```python
# Làm việc với mảng và tính toán số học
import numpy as np
```

```python
# Tính toán F1 score
from sklearn.metrics import f1_score
```

```python
# preds = [0.9 0.05 0.05 0 0 0]
# preds = [1 0 0 0 0 0 ]
```

* Sử dụng `np.argmax(preds, axis=1)` để tìm chỉ số của nhãn dự đoán có xác suất cao nhất cho mỗi mẫu.
* `.flatten()` được sử dụng để biến mảng 2D thành mảng 1D.
* Lặp qua từng nhãn để tính F1 score dựa trên các giá trị đã tính toán và trả về kết quả.

```python
# Hàm được định nghĩa để tính F1 score dựa trên các dự đoán (preds) và nhãn thực tế (labels)
def f1_score_func(preds, labels):
    preds_flat = np.argmax(preds, axis = 1).flatten()
    labels_flat = labels.flatten()
    return f1_score(labels_flat, preds_flat, average = 'weighted')
```

```python
#  Hàm để tính độ chính xác (accuracy) cho mỗi lớp (class) dựa trên các dự đoán và nhãn thực tế.
def accuracy_per_class(preds, labels):
    # Tạo một từ điển nghịch đảo (label_dict_inverse) để chuyển đổi giá trị nhãn (label) thành tên lớp.
    label_dict_inverse = {v: k for k, v in label_dict.items()}
    
    preds_flat = np.argmax(preds, axis = 1).flatten()
    labels_flat = labels.flatten()
    
    for label in np.unique(labels_flat):
        y_preds = preds_flat[labels_flat == label]
        y_true = labels_flat[labels_flat == label]
        print(f'Class: {label_dict_inverse[label]}')
        print(f'Accuracy: {len(y_preds[y_preds==label])}/{len(y_true)}\n')
```

## Task 9: Creating our Training Loop

Approach adapted from an older version of HuggingFace's `run_glue.py` script. Accessible [here](https://github.com/huggingface/transformers/blob/5bfcd0485ece086ebcbed2d008813037968a9e58/examples/run_glue.py#L128).

Thực hiện việc seed các giá trị ngẫu nhiên (random seed) để đảm bảo sự tái tạo của quá trình ngẫu nhiên trong Python. Việc này giúp đảm bảo các kết quả của việc huấn luyện mô hình có thể tái tạo được.

```python
import random

seed_val = 17
random.seed(seed_val)
np.random.seed(seed_val)
torch.manual_seed(seed_val)
torch.cuda.manual_seed_all(seed_val)
```

Kiểm tra xem GPU có sẵn để sử dụng hay không, và nếu có, thì bạn gán thiết bị là "cuda" (GPU), nếu không, bạn gán là "cpu" (CPU).

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

print(device)
```

```python
def evaluate(dataloader_val):

    model.eval()
    
    loss_val_total = 0
    predictions, true_vals = [], []
    
    # Lặp qua từng batch trong dataloader validation.
    for batch in tqdm(dataloader_val):
        
        # Đưa dữ liệu của mỗi batch lên thiết bị đã chọn.
        batch = tuple(b.to(device) for b in batch)
        
        # Đưa inputs qua mô hình để nhận được đầu ra.
        inputs = {'input_ids':      batch[0],
                  'attention_mask': batch[1],
                  'labels':         batch[2],
                 }
        
        # Sử dụng torch.no_grad() để tắt gradient computation, giúp tiết kiệm tài nguyên và tăng tốc độ.
        with torch.no_grad():        
            outputs = model(**inputs)
            
        loss = outputs[0]
        logits = outputs[1]
        loss_val_total += loss.item()

        logits = logits.detach().cpu().numpy()
        label_ids = inputs['labels'].cpu().numpy()
        predictions.append(logits)
        true_vals.append(label_ids)
    
    loss_val_avg = loss_val_total/len(dataloader_val) 
    
    predictions = np.concatenate(predictions, axis=0)
    true_vals = np.concatenate(true_vals, axis=0)
            
    return loss_val_avg, predictions, true_vals
```

```python
for epoch in tqdm(range(1, epochs+1)):
    
    model.train()
    
    # Khởi tạo tổng loss huấn luyện (loss_train_total) cho mỗi epoch.
    loss_train_total = 0
    
    # Sử dụng tqdm để theo dõi quá trình huấn luyện và tiến trình của từng batch.
    progress_bar = tqdm(dataloader_train,
                        desc = 'Epoch {:1d}'.format(epoch),
                        leave = False,
                        disable = False)
    for batch in progress_bar:
        
        model.zero_grad()
        
        batch = tuple(b.to(device) for b in batch)
        
        # Đưa dữ liệu của mỗi batch lên thiết bị đã chọn.
        inputs = {
            'input_ids'      : batch[0],
            'attention_mask' : batch[1],
            'labels'         : batch[2]
        }
        
        # Đưa dữ liệu vào mô hình để tính toán đầu ra (outputs).
        outputs = model(**inputs)
        
        # Thực hiện backpropagation và cập nhật gradient.
        loss = outputs[0]
        loss_train_total += loss.item()
        loss.backward()
        
        # Clip gradient để đảm bảo rằng gradient không quá lớn, giúp ổn định quá trình huấn luyện.
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        # Thực hiện bước tối ưu hóa (optimizer.step()) và cập nhật giá trị learning rate (scheduler.step()).
        optimizer.step()
        scheduler.step()
        
        # Cập nhật thông tin hiển thị trạng thái quá trình huấn luyện trên progress bar.
        progress_bar.set_postfix({'training_loss': '{:.3f}'.format(loss.item()/len(batch))})
        
    # Lưu mô hình sau mỗi epoch.
    torch.save(model.state_dict(), f'Model/BERT_ft_epoch{epoch}.model')
    
    tqdm.write('\nEpoch {epoch}')
    
    loss_train_avg = loss_train_total/len(dataloader)
    tqdm.write(f'Training loss: {loss_train_avg}')
    
    val_loss, predictions, true_vals = evaluate(dataloader_val)
    val_f1 = f1_score_func(predictions, true_vals)
    tqdm.write(f'Validation loss: {val_loss}')
    tqdm.write(f'F1 Score (weighted): {val_f1}')
```
